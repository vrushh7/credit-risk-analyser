## Step 3 — Preprocessing & Feature Engineering

```python
preprocessor = CreditRiskPreprocessor(test_size=0.2, random_state=42)
X_train, X_test, y_train, y_test = preprocessor.fit_transform(df)
feature_names = preprocessor.get_feature_names()

print(f'Features: {feature_names}')
print(f'X_train: {X_train.shape}, X_test: {X_test.shape}')
```

## Step 4 — Model Training

```python
fitted_models = train_models(X_train, y_train)
print('\n✅ All models trained and saved to models/')
```

## Step 5 — Evaluation

```python
results_df = evaluate_models(fitted_models, X_test, y_test)
results_df
```

```python
fig = plot_roc_curves(fitted_models, X_test, y_test)
plt.show()
```

```python
fig = plot_confusion_matrices(fitted_models, X_test, y_test)
plt.show()
```

```python
fig = plot_feature_importance(fitted_models, feature_names)
plt.show()
```

```python
fig = plot_metric_comparison(results_df)
plt.show()
```

## Step 6 — Business Insights

```python
best_name  = results_df['ROC-AUC'].idxmax()
best_model = fitted_models[best_name]
y_prob     = best_model.predict_proba(X_test)[:, 1]

print(f'Best model: {best_name}')
risk_df = print_risk_summary(y_prob, y_test)
risk_df.head()
```

```python
fig = plot_threshold_analysis(y_test, y_prob, model_name=best_name)
plt.show()
```

```python
importances = best_model.feature_importances_ if hasattr(best_model, 'feature_importances_') \
              else np.abs(best_model.coef_[0])
print_business_insights(feature_names, importances)
```
